```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
#test_size= 20%
#randon_state= hyper-parameter any +ve int is fine, 42 special-learn unit5
```

```python
x_train
```

```python
x_train=np.array(x_train).reshape(-1,1) #1D to 2D
x_train
```

```python
x_test=np.array(x_test).reshape(-1,1) #1D to 2D
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
m= lr.coef_
m
```

```python
b=lr.intercept_
b
```

```python
y_train_predict = (m*x_train)+b
y_train_predict.flatten()
```

```python
y_train_predict1= lr.predict(x_train)
y_train_predict1
```
