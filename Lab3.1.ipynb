```python
#Simple Linear Regression

from scipy import stats
import matplotlib.pyplot as plt

x=[2,4,6,8]
y=[3,7,5,10]

#b1=slope b0=intercept r=corelation(value b/w 0,1) >6 good/closely related p=hypothesis
b1,b0,r,p,std_err = stats.linregress(x,y)
print(b1,b0,r,p,std_err)

def myf(x):
    return b0+b1*x
mymodel = list(map(myf, x))
speed= myf(10)
plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()
print(speed)
```

```python
from scipy import stats
import matplotlib.pyplot as plt
import statsmodels.api as sm

x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]

#b1=slope b0=intercept r=corelation(value b/w 0,1) >6 good/closely related p=hypothesis
b1,b0,r,p,std_err = stats.linregress(x,y)
print(b1,b0,r,p,std_err)

def myf(x):
    return b0+b1*x
mymodel = list(map(myf, x))
speed= myf(20)
plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()
print(speed)
model= sm.OLS(y,x).fit()
print(model.summary())
```
