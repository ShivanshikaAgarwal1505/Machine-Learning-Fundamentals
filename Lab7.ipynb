```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("iris.xls")
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclassifier= DecisionTreeClassifier(criterion="entropy")
a= df.iloc[:, :-1]
b= df.iloc[:, -1]
dtclassifier.fit(a,b)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtclassifier, filled=True)
```

```python
plt.show()
```

```python
# Using Gini
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclassifier= DecisionTreeClassifier(criterion="gini")
a1= df.iloc[:, :-1]
b1= df.iloc[:, -1]
dtclassifier.fit(a1,b1)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtclassifier, filled=True)
```

```python
plt.show()
```

```python
# pruning
# pre- pruning= during trining, apply condition
# post-pruning= during testing, reduce/optimise the tree
```

```python
# post-Pruning using max_depth
from sklearn.tree import DecisionTreeClassifier
dtclassifier= DecisionTreeClassifier(criterion="gini", max_depth=2)
a1= df.iloc[:, :-1]
b1= df.iloc[:, -1]
dtclassifier.fit(a1,b1)
```

```python
from sklearn import tree
plt.figure(figsize=(5,5))
tree.plot_tree(dtclassifier, filled=True)
```

```python
plt.show()
```

```python
# for larger features, recommanded to do pre-pruning using Cross Validation
```

```python
from sklearn.model_selection import GridSearchCV
a2= df.iloc[:, :-1]
b2= df.iloc[:, -1]
```

```python
param_grid= {
    "criterion": ['gini', 'entropy'],
    "max_depth": [2,3,4,5],
    "min_samples_split": [2,3,4,5],
    "min_samples_leaf": [2,3,4,5]
}
```

```python
grid_search= GridSearchCV(dtclassifier, param_grid, cv=5)
```

```python
grid_search.fit(a2, b2)
```

```python
best_params=grid_search.best_params_
best_params
```

```python
best_classifier = grid_search.best_estimator_
best_classifier
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(best_classifier, filled=True)
```

```python
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(a2,b2, test_size=0.2)
```

```python
from sklearn import tree
DT= tree.DecisionTreeClassifier()
DT.fit(X_train, y_train)
y_pred= DT.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f'Model Accuracy Score: :, {100*accuracy_score(y_test, y_pred):0.2f}')
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
print('Confusion Matrix')
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
